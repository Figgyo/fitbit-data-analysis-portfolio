```python
hourly_steps = dfs['cleaned_hourlySteps'].copy()

# Convert time column to datetime
hourly_steps['ActivityHour'] = pd.to_datetime(hourly_steps['ActivityHour'])

# Extract hour and day of week
hourly_steps['Hour'] = hourly_steps['ActivityHour'].dt.hour
hourly_steps['DayOfWeek'] = hourly_steps['ActivityHour'].dt.day_name()
hourly_steps['IsWeekend'] = hourly_steps['DayOfWeek'].isin(['Saturday', 'Sunday'])
```

```python
import matplotlib.pyplot as plt

avg_steps_hour = hourly_steps.groupby('Hour')['StepTotal'].mean()

avg_steps_hour.plot(kind='line', marker='o', title='Average Steps by Hour of Day')
plt.xlabel('Hour (0–23)')
plt.ylabel('Average Steps')
plt.grid(True)
plt.show()
```

```python
avg_steps_by_type = hourly_steps.groupby(['Hour', 'IsWeekend'])['StepTotal'].mean().unstack()

avg_steps_by_type.plot(kind='line', marker='o', title='Weekday vs Weekend: Steps by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Steps')
plt.legend(['Weekday', 'Weekend'])
plt.grid(True)
plt.show()
```

```python
hourly_intensity = dfs['cleaned_hourlyIntensities'].copy()
hourly_calories = dfs['cleaned_hourlyCalories'].copy()

# Convert to datetime
hourly_intensity['ActivityHour'] = pd.to_datetime(hourly_intensity['ActivityHour'])
hourly_calories['ActivityHour'] = pd.to_datetime(hourly_calories['ActivityHour'])

# Extract hour, day of week, and weekend indicator
for df in [hourly_intensity, hourly_calories]:
    df['Hour'] = df['ActivityHour'].dt.hour
    df['DayOfWeek'] = df['ActivityHour'].dt.day_name()
    df['IsWeekend'] = df['DayOfWeek'].isin(['Saturday', 'Sunday'])
```

```python
import matplotlib.pyplot as plt

# Plot average 'TotalIntensity' by hour
avg_intensity_by_type = hourly_intensity.groupby(['Hour', 'IsWeekend'])['TotalIntensity'].mean().unstack()

avg_intensity_by_type.plot(kind='line', marker='o', title='Weekday vs Weekend: Intensity by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Intensity')
plt.legend(['Weekday', 'Weekend'])
plt.grid(True)
plt.show()
```

```python
# Plot average 'Calories' by hour
avg_calories_by_type = hourly_calories.groupby(['Hour', 'IsWeekend'])['Calories'].mean().unstack()

avg_calories_by_type.plot(kind='line', marker='o', title='Weekday vs Weekend: Calories Burned by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Calories Burned')
plt.legend(['Weekday', 'Weekend'])
plt.grid(True)
plt.show()
```

```python
# Load daily activity
daily = dfs['cleaned_dailyActivity']

# Calculate total active minutes per day (sum of activity columns)
active_cols = ['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes']
daily['TotalActiveMinutes'] = daily[active_cols].sum(axis=1)

# Calculate proportion of sedentary vs. active time
daily['TotalMinutes'] = daily['TotalActiveMinutes'] + daily['SedentaryMinutes']
daily['SedentaryPercent'] = daily['SedentaryMinutes'] / daily['TotalMinutes'] * 100
daily['ActivePercent'] = daily['TotalActiveMinutes'] / daily['TotalMinutes'] * 100

# Boxplot to visualize
import seaborn as sns
sns.boxplot(data=daily[['SedentaryPercent', 'ActivePercent']])
plt.title('Distribution of Sedentary vs. Active Percent of Waking Time')
plt.ylabel('Percent of Waking Time')
plt.show()

# Average by day of week
daily['ActivityDate'] = pd.to_datetime(daily['ActivityDate'])
daily['DayOfWeek'] = daily['ActivityDate'].dt.day_name()
avg_sedentary = daily.groupby('DayOfWeek')['SedentaryPercent'].mean().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
avg_sedentary.plot(kind='bar', title='Average Sedentary Percent by Day of Week')
plt.ylabel('Average Sedentary Percent (%)')
plt.show()
```

```python
# Mark each day as "streak success" or not
daily['Reached_10k'] = daily['TotalSteps'] >= 10000

# Sort by user and date
daily = daily.sort_values(['Id', 'ActivityDate'])

# Calculate streaks per user
def longest_streak(arr):
    max_streak = streak = 0
    for reached in arr:
        if reached:
            streak += 1
            max_streak = max(max_streak, streak)
        else:
            streak = 0
    return max_streak

streaks = daily.groupby('Id')['Reached_10k'].apply(longest_streak)

# Plot distribution of longest streaks
streaks.plot(kind='hist', bins=range(1, streaks.max()+2), rwidth=0.8)
plt.title('Distribution of Longest Step Goal Streaks per User')
plt.xlabel('Longest Consecutive Days Achieving 10,000 Steps')
plt.ylabel('Number of Users')
plt.show()

print(f"Median streak length: {streaks.median()} days")
print(f"Max streak length: {streaks.max()} days")
```
